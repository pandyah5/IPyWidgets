## Dynamic plotting

```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual

def f(omega):
    x = np.linspace(-20, 20, 20000)
    y = np.sin(omega*x)
    plt.plot(x, y, label = "Sin(mx)")
    ax = plt.gca()
    ax.set_ylim([-2, 2])
    ax.set_xlim([-2, 2])
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend(loc = 'best')
    plt.show()
    

interact(f, omega = 10);
```

## Dynamic plotting with string manipulation with sympy
#### Program that plots the function and it's derivative

```python
import matplotlib.pyplot as plt
import numpy as np 
import sympy as sp
from sympy import *
import IPython.display as display
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual

print('''Input should be given in this format: 1*x**2 + 1*x**1 + 1*x**0''')
a = input("Enter the polynomial here: ")
def calculus(ans_1, ans_2, ans_3):
    if ans_1 == 'Yes':
        x = np.linspace(-100,100,20000)
        b = a.split('+')
        p = []
        c = []
        n = 0
        z = 0
        m = 0
        fun = 0
        while z < len(b):
            c.append(b[z].split('*')[0])
            z += 1
        while n < len(b):
            p.append(b[n].split('**')[1])
            n += 1
        while m < len(p):
            fun += float(c[m])*x**float(p[m]) 
            m += 1
        plt.plot(x, fun, label = 'f(x)', color = 'red')
        plt.plot(x, 0*x, color = 'black')
        plt.plot(0*x, x, color = 'black')
        ax = plt.gca()
        ax.set_xlim(-10, 10)
        ax.set_ylim(-25, 25)
        plt.xlabel('x')
        plt.ylabel('Y')
        plt.title('Function Graph')
        plt.legend()
        plt.grid()
        plt.show()
    else:
        pass
    
    if ans_2 == 'Yes':
        x = Symbol('x')
        b = a.split('+')
        p = []
        c = []
        n = 0
        z = 0
        m = 0
        fun = 0
        while z < len(b):
            c.append(b[z].split('*')[0])
            z += 1
        while n < len(b):
            p.append(b[n].split('**')[1])
            n += 1
        while m < len(p):
            fun += float(c[m])*x**float(p[m]) 
            m += 1
        fun_prime = fun.diff(x)
        print("The differentiated function is: ",fun_prime)

    else:
        pass
    
    if ans_3 == 'Yes':
        x = Symbol('x')
        b = a.split('+')
        p = []
        c = []
        n = 0
        z = 0
        m = 0
        fun = 0
        while z < len(b):
            c.append(b[z].split('*')[0])
            z += 1
        while n < len(b):
            p.append(b[n].split('**')[1])
            n += 1
        while m < len(p):
            fun += float(c[m])*x**float(p[m]) 
            m += 1
        fun_prime = fun.diff(x)
        ## Trying to graph f'(x)
        x1 = np.linspace(-100, 100, 20000)
        y = str(fun_prime)
        d = y.split('+')
        c_1 = 0
        constant = []
        while c_1 < len(d):
            if 'x' in d[c_1]:
                c_1 += 1
            else:
                constant.append(d[c_1])
                c_1 += 1
        c_2 = 0
        p_1 = 0
        f = 0
        func = 0
        constant1 = []
        power = []
        while c_2 < len(d):
            if '*' in d[c_2]:
                constant1.append(d[c_2].split('*')[0])
                c_2 += 1
            else:
                c_2 += 1
        while p_1 < len(d):
            if '**' in d[p_1]:
                power.append(d[p_1].split('**')[1])
                p_1 += 1
            else:
                p_1 += 1
        while f < len(power):
            func += float(constant1[f])*x1**float(power[f])
            f += 1
        
        func += float(constant[0])
        plt.plot(x1, func, color = 'Blue', label = "f'(x)")
        plt.plot(x1, 0*x1, color = 'black')
        plt.plot(0*x1, x1, color = 'black')
        ax = plt.gca()
        ax.set_xlim(-10, 10)
        ax.set_ylim(-25, 25)
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title('Derivative Graph')
        plt.legend()
        plt.grid()
        plt.show()

style = {'description_width': 'initial'}
interact(calculus,
        ans_1 = widgets.RadioButtons(options = ['Yes', 'No'], 
                disabled = False, description = "Do you want to plot your function?", style = style),
        ans_2 = widgets.RadioButtons(options = ['Yes', 'No'], 
                disabled = False, description = "Do you want to differentiate your function?", style = style),
        ans_3 = widgets.RadioButtons(options = ['Yes', 'No'], 
                disabled = False, description = "Do you want to plot the derivative of f(x)?", style = style))
```

## Multiple Dynamic plots
#### Electronic Configuration of first 10 elements

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display
import ipywidgets as widgets 
from ipywidgets import interact, interactive, fixed, interact_manual

def orbital(element):
    """
    This function helps in plotting bohr's atomic model for the first ten elements 
    with the use of interactive widgets and makes learning enjoyable!!
    """
    At_num = {'Hydrogen' : 1, 
              'Helium' : 2, 
              'Lithium' : 3, 
              'Beryllium' : 4, 
              'Boron' :5, 
              'Carbon' : 6,
              'Nitrogen' : 7, 
              'Oxygen' :8, 
              'Fluorine' :9, 
              'Neon' : 10}
    e = At_num[element]
    r0 = 5
    ## Writing code for hydrogen atom
    if e == 1:
        s = np.linspace(0, 2 * np.pi, 1000)
        x = r0*np.cos(s)
        y = r0*np.sin(s)
        z = [2, 5]
        p0_x = r0*np.cos(z[0])
        p0_y = r0*np.sin(z[0])
        plt.plot(p0_x, p0_y, 'ro', color = 'red', label = 'Electrons')
        plt.plot(x, y, label = '1st Orbit', color = 'black')
        plt.plot(0,0, 'ro', color = 'cyan', label = 'Nucleus')
        plt.title('Electronic configuration of hydrogen')
        plt.legend(loc = 1)
        plt.show()
    ## Code for helium
    elif e == 2:
        s = np.linspace(0, 2 * np.pi, 1000)
        x = r0*np.cos(s)
        y = r0*np.sin(s)
        z = [2, 5]
        p1_x = r0*np.cos(z[0])
        p1_y = r0*np.sin(z[0])
        p2_x = r0*np.cos(z[1])
        p2_y = r0*np.sin(z[1])
        plt.plot(p1_x, p1_y, 'ro', color = 'red', label = 'Electrons')
        plt.plot(p2_x, p2_y, 'ro', color = 'red')
        plt.plot(0,0, 'ro', color = 'cyan', label = 'Nucleus')
        plt.plot(x, y, label = '1st Orbit', color = 'black')
        plt.title('Electronic configuration of helium')
        plt.legend(loc = 1)
        plt.show()
    ## Code for the rest of the elements
    else:
        r0 = 5
        r1 = 7
        i = 0
        a = [0]
        b = [0]
        s = np.linspace(0, 2 * np.pi, 1000)
        x1 = r1*np.cos(s)
        y1 = r1*np.sin(s)
        x = r0*np.cos(s)
        y = r0*np.sin(s)
        ## Setting up loops to reduce code length
        while i <= 1:
            a = a + [r0*np.cos(i)]
            b = b + [r0*np.sin(i)]
            i += 1
        j = 3
        c = [0]
        d = [0]
        while j >= 3 and j <= e:
            c = c + [r1*np.cos(j)]
            d = d + [r1*np.sin(j)]
            j += 1
        plt.plot(a, b, 'ro', color = 'red', label = 'Electrons')
        plt.plot(c, d, 'ro', color = 'red')
        plt.plot(x1, y1, label = '2nd Orbit', color = 'green')
        plt.plot(x, y, label = '1st Orbit', color = 'black')
        plt.plot(0, 0, 'ro', color = 'cyan', label = 'Nucleus')
        plt.title('Electronic configuration of ' + str(element))
        plt.legend(loc = 1)
        plt.show()
## Setting up widgets for controlling the function         
interact(orbital, 
        element = widgets.Dropdown(
    options=['Hydrogen', 'Helium', 'Lithium','Beryllium','Boron','Carbon','Nitrogen','Oxygen','Fluorine','Neon'],
    value='Hydrogen',
    description='Element:',
    disabled=False,
));
```

#### Plotting multiples types of parametric functions

```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display
from ipywidgets import interact, interactive, fixed, interact_manual

def plot(curve, slope, cst, freq, amp, phase, rad):
    if curve == 'Line':
        x = np.linspace(-50,50,20000)
        y = slope*x + cst 
        plt.plot(x, y, label = 'y = m*x + c', color = 'green')
        ax = plt.gca()
        ax.set_xlim([-40, 40])
        ax.set_ylim([-10, 50])
        plt.legend(loc = 'best')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title('Line Plot')
        plt.show()
    elif curve == 'Parabola':
        x = np.linspace(-50,50,20000)
        y = cst*x**2
        plt.plot(x, y, label = 'y = c*x^2', color = 'green')
        ax = plt.gca()
        ax.set_xlim([-40, 40])
        ax.set_ylim([-10, 50])
        plt.legend(loc = 'best')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title('Parabola')
        plt.show()
    elif curve == 'Sine curve':
        x = np.linspace(-50,50,20000)
        y = amp*np.sin(2*np.pi*freq*x + 5*phase) + cst 
        plt.plot(x, y, label = 'y = Sin(x)', color = 'green')
        ax = plt.gca()
        ax.set_xlim([-40, 40])
        ax.set_ylim([-10, 50])
        plt.legend(loc = 'best')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title('Sine curve')
        plt.show()
    elif curve == 'Cosine curve':
        x = np.linspace(-50,50,20000)
        y = amp*np.cos(2*np.pi*freq*x + 5*phase) + cst 
        plt.plot(x, y, label = 'y = cos(x)', color = 'green')
        ax = plt.gca()
        ax.set_xlim([-40, 40])
        ax.set_ylim([-10, 50])
        plt.legend(loc = 'best')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title('Cosine curve')
        plt.show()
    elif curve == 'Tangent Curve':
        x = np.linspace(-50,50,20000)
        y = amp*np.tan(2*np.pi*freq*x + 5*phase) + cst 
        plt.plot(x, y, label = 'y = tan(x)', color = 'green')
        ax = plt.gca()
        ax.set_xlim([-40, 40])
        ax.set_ylim([-10, 50])
        plt.legend(loc = 'best')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title('Tangent curve')
        plt.show()
    elif curve == 'Circle':
        s = np.linspace(-50,50,20000)
        x = rad*np.cos(s)
        y = rad*np.sin(s)
        plt.plot(x, y, label = 'y^2 + x^2  = r^2', color = 'green')
        ax = plt.gca()
        ax.set_xlim([-40, 40])
        ax.set_ylim([-10, 50])
        plt.legend(loc = 'best')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title('Circle')
        plt.show()
    else:
        pass
    
interact(plot, 
        curve = widgets.Dropdown(options = ['Line', 'Parabola', 'Sine curve', 'Cosine curve', 'Tangent Curve', 'Circle'], description = 'Type of curves'),
        slope = widgets.FloatSlider(min = 0.0, max = 15.0, description = 'Slope*'), 
        cst = widgets.FloatSlider(min = 0.0, max = 5.0, description = 'Constant**'),
        rad = widgets.FloatSlider(min = 2.0, max = 10.0, description = 'Radius****'),
        freq = widgets.FloatSlider(min = 0.0, max = 1.0, description = 'Frequency***'),
        amp = widgets.FloatSlider(min = 0.0, max = 10.0, description = 'Amplitude***'),
        phase = widgets.FloatSlider(min = 0.0, max = 15.0, description = 'Phase***'))

print(' * :- Valid only for line plots', '\n', '** :- Valid for all', '\n', '*** :- Valid only for trigonometric functions ','\n','**** :- Valid only for circle')
```
